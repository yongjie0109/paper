```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def analyze_sequence(max_n):

    # N = [2 3 4 5 ... ]
    n_values = np.arange(2, max_n + 1)
    
    # C_N = (N - 1) / (2N - 1)
    c_n = (n_values - 1) / (2 * n_values - 1)
    
    # val = 2*Cn - Cn^(1/N)
    values = 2 * c_n - np.power(c_n, 1 / n_values)
    
    # 檢查是否有任何值小於 0
    min_val = np.min(values)
    is_negative = np.any(values < 0)
    
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, values, label=r'$2C_N - C_N^{1/N}$', color='blue', linewidth=1.5)
    
    # 繪製 y=0 的參考線
    plt.axhline(0, color='red', linestyle='--', linewidth=1, label='y=0')
    
    plt.title(f'Convergence of $2C_N - C_N^{{1/N}}$ (N up to {max_n})')
    plt.xlabel('N')
    plt.ylabel('Value')
    plt.grid(True, which='both', linestyle='--', alpha=0.6)
    plt.legend()
    
    # 顯示幾個數值以便觀察
    print(f"--- 分析結果 (Max N = {max_n}) ---")
    print(f"N = 2      : {values[0]:.8f}")
    print(f"N = 10     : {values[8]:.8f}")
    print(f"N = 100    : {values[98]:.8f}")
    print(f"N = 10000  : {values[9998]:.8f}")
    print(f"N = {max_n}: {values[-1]:.8f}")
    print("-" * 30)
    
    if is_negative:
        print("結論: 發現了小於 0 的數值。")
    else:
        print(f"結論: 在測試範圍內，所有數值皆大於 0。")
        print(f"最小值發生在 N={n_values[np.argmin(values)]}，值為 {min_val:.8f}")
        print("數值呈現趨近於 0 但保持正值的趨勢。")

    plt.show()

# N_max
analyze_sequence(1000000)
```
